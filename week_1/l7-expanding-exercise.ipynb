## Exercise 1 — Reply to a Negative Review

**Customer review** — store in a variable called `review`:

> I bought the 12-cup air fryer during your November sale for $59, but by December the price dropped to $39 on the same listing. The basket coating started peeling after six weeks of normal use. I called support and was told the warranty only covered manufacturing defects, not wear. The unit still works, but I feel misled by the pricing and expected better durability for a kitchen appliance.

**Sentiment:** `negative` (store in a variable called `sentiment`)

**Your task:**

Write a prompt where the model acts as a **customer service AI assistant** and generates an email reply. Requirements:

- Use the customer review (delimited by triple backticks) and the provided sentiment
- If sentiment is **negative**, apologize and suggest contacting customer service
- If sentiment is **positive** or **neutral**, thank the customer
- Include **specific details** from the review
- Write in a **concise, professional** tone
- Sign the email as `AI customer agent`
- Use `temperature=0` (default)

Print the response.
