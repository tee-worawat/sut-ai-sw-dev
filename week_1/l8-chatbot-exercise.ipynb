## Exercise 4 — Build a Simple OrderBot (Text Only)

The lesson built **OrderBot** for a pizza restaurant. You will build a simpler **CafeBot** using messages only (no Panel GUI).

```
Drinks:
- espresso 2.50
- latte 3.75 (small), 4.50 (large)
- iced tea 3.00

Pastries:
- croissant 2.80
- blueberry muffin 3.20

Extras:
- oat milk 0.60
- extra shot 0.80
```

**Your task:**

2. Start `context` with that system message. Send these user messages **one at a time**, appending each assistant reply to `context`:
   - "Hi, I'd like a large latte with oat milk and a croissant."
   - "Pickup please."
   - "That's all."

3. Print the assistant's response after each user message.
