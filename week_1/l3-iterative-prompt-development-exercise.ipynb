## Exercise 1 — Baseline Description

**Your task:**

Write your **first draft** prompt that asks the model to create a retail website product description based on the fact sheet. Use triple backticks as delimiters around the technical specifications.

Do **not** add length limits or audience constraints yet — this is your starting point.

Print the response and note: Is it too long? Too generic? Missing key details?

---

## Exercise 2 — Issue: Text Is Too Long

**Your task:**

Refine your prompt from Exercise 1. Add a constraint so the description uses **at most 50 words**.

Print the response. Optionally verify the word count in a second line of code.

---

## Exercise 3 — Issue: Wrong Focus for the Audience

**Your task:**

Refine your prompt again. The description is now for **furniture retailers**, not end consumers. Instruct the model to:

- Keep the **50-word limit**
- Write in a **technical** tone focused on **materials and construction**

Print the response.

---

## Exercise 4 — Add Product IDs

**Your task:**

Further refine your prompt. In addition to the requirements from Exercise 3, instruct the model to **include every 7-character Product ID** from the fact sheet at the end of the description.

Print the response and confirm both product IDs appear.

---

## Exercise 5 — Add a Dimensions Table (HTML)

**Your task:**

Create your **final** prompt. Build on Exercise 4 and also require:

1. Output formatted as **HTML** suitable for a website
2. Place the description inside a `<div>` element
3. After the description, include a table titled **Product Dimensions** with two columns: dimension name and measurement **in inches only**

Print the response. Optionally display it with `IPython.display.HTML`.
