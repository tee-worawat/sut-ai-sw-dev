## Exercise 6 — Work Out Your Own Solution First (Principle 2)

A student was asked to write a cost function for a small event venue. Review the problem and the student's answer below.

**Question:**

> I am renting a hall for a one-day conference. The venue charges $500 flat for the day, plus $15 per attendee for catering, plus $8 per attendee for audio-visual equipment. What is the total cost as a function of the number of attendees?

**Student's solution:**

> Let n be the number of attendees.
> 1. Venue fee: 500
> 2. Catering: 15n
> 3. A/V equipment: 15n
> Total cost: 500 + 15n + 15n = 500 + 30n

**Your task:**

Write a prompt that instructs the model to:

1. **First** solve the problem itself (showing steps and the final formula)
2. **Then** compare its answer to the student's solution
3. **Only after that**, decide whether the student is correct or incorrect

Use delimiters and a **fixed response format** with these sections:

```
Question:
```
...question here...
```

Student's solution:
```
...student work here...
```

Actual solution:
```
...your worked solution...
```

Is the student's solution the same as the actual solution:
```
yes or no
```

Student grade:
```
correct or incorrect
```

Print the response. *(Hint: the student made an error — your prompt structure should help the model catch it.)*

**Tactic practiced:** Instruct the model to work out its own solution before rushing to a conclusion.

---

## Bonus (Optional) — Hallucinations

The lesson ended with a note on **hallucinations**: models can sound confident about facts that are not real.

**Your task:**

Ask the model to tell you about a product that **does not exist**:

> **NebulaSound WavePro 9000** wireless earbuds by **AudioForge Labs**

Run the query once with a plain prompt. Then revise your prompt to instruct the model to **say when it is unsure whether a product is real** and to **avoid inventing specifications**. Compare the two responses.

Write both prompts and print both responses.
