## ✅ Your Pydantic model should include:

- ✓ A class that inherits from `BaseModel`
- ✓ Field types matching your expected LLM output structure
- ✓ Validation constraints (string length, numeric ranges, allowed choices)
- ✓ Required fields (without defaults) and optional fields (with `Optional[]` and default values)
- ✓ Field() constraints like `ge=`, `le=`, `min_length=`, etc.

---

<details>
<summary><strong>📚 Refresher: Understanding Pydantic Models (click to expand)</strong></summary>

A Pydantic model inherits from `BaseModel` and defines:
1. **Field names**: Python variable names (e.g., `title`, `genre`)
2. **Field types**: Python types (e.g., `str`, `int`, `float`, `bool`)
3. **Constraints**: Validation rules using `Field()` (e.g., `ge=0`, `min_length=1`)
4. **Optional fields**: Fields that can be `None` using `Optional[type] = None`

**Example structure:**
```python
class MyModel(BaseModel):
    required_field: str
    number_field: int = Field(..., ge=0, le=100)
    optional_field: Optional[str] = None
    choice_field: Literal["option1", "option2", "option3"]
```

</details>

<details>
<summary><strong>📚 Refresher: Common Field Types & Constraints (click to expand)</strong></summary>

**Basic types:**
- `str`: String
- `int`: Integer
- `float`: Float
- `bool`: Boolean

**Pydantic types:**
- `Literal["opt1", "opt2"]`: Must be one of specific values
- `List[str]`: List of strings
- `Optional[type]`: Can be None

**Common constraints:**
- `min_length`, `max_length`: String/list length
- `ge`, `le`: Greater/less than or equal (numbers)
- `gt`, `lt`: Greater/less than (numbers)
- `description`: Documentation string

</details>

<details>
<summary><strong>📚 Refresher: Structured Output with Gemini (click to expand)</strong></summary>

The Gemini API supports requesting structured output that matches a Pydantic model. Instead of getting unstructured text, the LLM returns JSON that conforms to your schema.

**How it works:**
1. You define a Pydantic model (already done in Step 1!)
2. You pass the model schema to the API using `response_json_schema`
3. The LLM returns structured JSON matching your model
4. You validate the response to ensure it meets all constraints

**Key benefits:**
- LLM output is more predictable and structured
- Easier to validate and use in your application
- Catches format errors before they cause bugs

**Note:** Even with structured output, validation is still important! The LLM might return data that fits the schema but violates your constraints (e.g., a rating of 15.0 when max is 10.0).

</details>

<details>
<summary><strong>📚 Refresher: Making LLM API Calls (click to expand)</strong></summary>

**Basic Gemini API call with structured output:**
```python
from gemini_helpers import get_structured_output

class MyResponse(BaseModel):
    field1: str
    field2: int

response = get_structured_output(
    prompt="User query here",
    response_model=MyResponse,
    system_instruction="You are a helpful assistant.",
)
```

**The response object** is a Python dictionary that you can validate with your Pydantic model.

</details>

<details>
<summary><strong>📚 Refresher: Validation Workflow with LLM Responses (click to expand)</strong></summary>

1. **Create user queries** as strings (from your example prompts)
2. **Call the LLM** using `get_structured_output()` with your Pydantic model
3. **Get structured response** as a validated Pydantic instance
4. **Validate response** using try/except to catch `ValidationError`
5. **Handle errors** gracefully—print what went wrong

**Why validate LLM output?**
- LLMs can hallucinate values outside your constraints
- Ensures data quality before using in your application
- Provides clear error messages when LLM returns invalid data
- Catches edge cases (negative ratings, invalid dates, etc.)

**Example validation:**
```python
try:
    validated = MyModel(**response)
    print("✅ Valid response:", validated.model_dump_json(indent=2))
except ValidationError as e:
    print("❌ Validation error:", e)
```
